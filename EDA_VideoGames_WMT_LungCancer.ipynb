<p>3. Lognormal distribution model</p>

```python
def MakeNormalPlot(years):
    """Generates a normal probability plot of birth weights.

    weights: sequence
    """
    mean, var = thinkstats2.TrimmedMeanVar(years, p=0.01)
    std = np.sqrt(var)

    xs = [-5, 5]
    xs, ys = thinkstats2.FitLine(xs, mean, std)
    thinkplot.Plot(xs, ys, color='0.8', label='model')

    xs, ys = thinkstats2.NormalProbability(years)
    thinkplot.Plot(xs, ys, label='years')
```

```python
log_weights = np.log10(games['Year_of_Release'])
MakeNormalPlot(log_weights)
thinkplot.Config(title='Year of Release, lognormal plot', xlabel='Years (log10)',
                 ylabel='CDF', loc='upper left')
```

<p>The CDF of the lognormal distribution is the same as the CDF of the normal distribution, with log x substituted for x.</p>

# Linear Least Squares

We are decided to take from our sample columns: 'Global Sales' and 'Critic Score'. Because in our opinion they should be related to each other. As Global sales higher then critic score should be higher too.

```python
from thinkstats2 import Mean, MeanVar, Var, Std, Cov

def LeastSquares(xs, ys):
    meanx, varx = MeanVar(xs)
    meany = Mean(ys)

    slope = Cov(xs, ys, meanx, meany) / varx
    inter = meany - slope * meanx

    return inter, slope
```

```python
inter, slope = LeastSquares(game['Global_Sales'], game['Critic_Score'])
inter, slope
```

Here we defined interception and slope of our linear model. Now, we are going to find Residuals:

```python
def Residuals(xs, ys, inter, slope):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    res = ys - (inter + slope * xs)
    return res
```

```python
Residuals(game['Global_Sales'], game['Critic_Score'], inter, slope)
```

Now, to test our results, we will use FitLine to understand relation them between Linear model

```python
def FitLine(xs, inter, slope):
    fit_xs = np.sort(xs)
    fit_ys = inter + slope * fit_xs
    return fit_xs, fit_ys
```

```python
fit_xs, fit_ys = FitLine(game['Global_Sales'], inter, slope)
```

```python
def GroupByDay(walmart, func=np.mean):
    groups = walmart[['Date', 'High']].groupby('Date')
    daily = groups.aggregate(func)
    daily['Date'] = daily.index
    start = daily.Date[0]
    one_year = np.timedelta64(1, 'Y')
    daily['years'] = (daily.Date - start) / one_year
    return daily
```

Now, by this CDF let's plot SF

```python
import survival

def MakeSurvivalFromCdf(cdf, label=''):
    """Makes a survival function based on a CDF.

    cdf: Cdf
    
    returns: SurvivalFunction
    """
    ts = cdf.xs
    ss = 1 - cdf.ps
    return survival.SurvivalFunction(ts, ss, label)
```

```python
sf_cancer = MakeSurvivalFromCdf(cdf_cancer, label='survival curve')
```

```python
thinkplot.Plot(sf_cancer)
thinkplot.Cdf(cdf_cancer, alpha=0.2)
thinkplot.Config(loc='center left')
```

And Hazard Function

```python
hf_cancer = sf_cancer.MakeHazardFunction(label='hazard')
thinkplot.Plot(hf_cancer)
thinkplot.Config(loc='upper left')
```

Now Kaplan-Meier function

```python
from collections import Counter

def EstimateHazardFunction(complete, ongoing, label='', verbose=False):
    """Estimates the hazard function by Kaplan-Meier.

    http://en.wikipedia.org/wiki/Kaplan%E2%80%93Meier_estimator

    complete: list of complete lifetimes
    ongoing: list of ongoing lifetimes
    label: string
    verbose: whether to display intermediate results
    """
    if np.sum(np.isnan(complete)):
        raise ValueError("complete contains NaNs")
    if np.sum(np.isnan(ongoing)):
        raise ValueError("ongoing contains NaNs")

    hist_complete = Counter(complete)
    hist_ongoing = Counter(ongoing)

    ts = list(hist_complete | hist_ongoing)
    ts.sort()

    at_risk = len(complete) + len(ongoing)

    lams = pd.Series(index=ts)
    for t in ts:
        ended = hist_complete[t]
        censored = hist_ongoing[t]

        lams[t] = ended / at_risk
        if verbose:
            print(t, at_risk, ended, censored, lams[t])
        at_risk -= ended + censored

    return survival.HazardFunction(lams, label=label)
```

```python
hf = EstimateHazardFunction(walmart['High'], walmart['Open'])
thinkplot.Plot(hf)
thinkplot.Config(ylim=[0, 0.1], xlabel='Price (High)',
                 ylabel='Hazard')
```

Here we have taken such features as High (highest price for the product at the current date) and "Open"(the price set for the product at the opening of the store) to estimate the Hazard Function.
